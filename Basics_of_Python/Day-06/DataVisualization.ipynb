```python
# Histogram
import matplotlib.pyplot as plt
```

```python
df.Make.value_counts().nlargest(40).plot(kind='bar',figsize=(10,5))
plt.title('Number of car by make')
plt.ylabel('Number of cars')
plt.xlabel('Make')
```

```python
df.Make.value_counts().nlargest(40).plot(kind='pie',figsize=(10,5))
plt.title('Number of car by make')
plt.ylabel('Number of cars')
plt.xlabel('Make')
```

```python
df.Make.value_counts().nlargest(40).plot(kind='chart',figsize=(10,5))
plt.title('Number of car by make')
plt.ylabel('Number of cars')
plt.xlabel('Make')
```
